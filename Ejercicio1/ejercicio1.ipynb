```python
import boto3
```

```python
client = boto3.client('s3')
bucket_name = 'my-bucket-dbt99'
```

```python
# Create a new S3 bucket
response = client.create_bucket(
    Bucket=bucket_name,
    CreateBucketConfiguration={
        'LocationConstraint': 'us-west-2'
    }
    )
print(response)
```

```python
# Upload a file to the bucket
file_name = 'example.txt'

response = client.upload_file(Bucket=bucket_name, Key=file_name, Filename=f'./{file_name}')
print(response)
```

```python
# Get files in the bucket
response = client.list_objects_v2(Bucket=bucket_name)
print(response)


for i, obj in enumerate(response.get('Contents', [])):
    print(f"Object {i+1}: {obj['Key']}") # Prints the file path/name
```

```python
# Download a file from the bucket
response = client.download_file(Bucket=bucket_name, Key=file_name, Filename=f'./downloaded_{file_name}')
print(response)
```
